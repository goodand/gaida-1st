```python
# 컬럼들 진짜로 삭제

# 진짜로 컬럼 삭제한 df 리턴
df.drop(['부가기호','Unnamed: 13'],axis=1)

# 리턴 말고 원본에서 삭제하기
df.drop(['부가기호','Unnamed: 13','주제분류번호'],axis=1,inplace=True) # 이미 없어져서 오류가 난듯 # KeyError: "['부가기호', 'Unnamed: 13'] not found in axis"
```

```python
# Col 중에 하나라도 NaN(비어 있음)이면 그 Col 삭제
book_df = df.dropna(axis=1)

book_df.head(3)

# col의 모든 데이터가 NaN아묜 햐덩 Col 삭제
book_df = df.dropna(axis=1,how='all')
book_df.head(3)

## 위에꺼 복사본이라 괜춘
```
